```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('datasets/insurance_data.csv')
df.head()
```

```python
plt.xlabel('Age')
plt.ylabel('Bought Insurance?')
plt.scatter(df['age'], df['bought_insurance'], color='blue', marker='+')
```

```python
X = df['age']
y = df['bought_insurance']
X.size, y.size
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(pd.DataFrame(df.loc[:, 'age']) , pd.DataFrame(df.loc[:, 'bought_insurance']) , test_size=0.1)
X_train.size, y_train.size
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test, y_test)
```
