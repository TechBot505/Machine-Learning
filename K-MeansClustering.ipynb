# K Means Clustering Algorithm
* K-means clustering is a simple and elegant approach for partitioning a data set into K distinct, non-overlapping clusters.
* To perform K-means clustering, we must first specify the desired number of clusters K; then the K-means algorithm will assign each observation to exactly one of the K clusters.
* The K-means algorithm aims to choose centroid that minimise the **inertia**, or **within-cluster sum-of-squares criterion**.

[K Means Clustering Medium Blog](https://medium.com/data-folks-indonesia/step-by-step-to-understanding-k-means-clustering-and-implementation-with-sklearn-b55803f519d6)

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('datasets/income.csv')
df.head()
```

```python
plt.scatter(df['Age'], df['Income($)'], color='blue', marker='*')
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
model = KMeans(n_clusters=3)
model
```

```python
y_pred = model.fit_predict(df[['Age', 'Income($)']])
```

```python
df['cluster'] = y_pred
df
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='purple')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
model2 = KMeans(n_clusters=3)
y_pred = model2.fit_predict(df[['Age', 'Income($)']])
y_pred
```

```python
df
```

```python
df['cluster2'] = y_pred
df
```

```python
df1 = df[df.cluster2 == 0]
df2 = df[df.cluster2 == 1]
df3 = df[df.cluster2 == 2]
plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='purple')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```
